```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts.few_shot import FewShotChatMessagePromptTemplate
from langchain.callbacks import StreamingStdOutCallbackHandler
from langchain.prompts import ChatMessagePromptTemplate, ChatPromptTemplate

llm = ChatOpenAI(
    temperature=0.1,
    model="gpt-4o-mini",
    streaming=True,
    callbacks=[StreamingStdOutCallbackHandler()]
)

examples = [
    {
        "movie": "Inception",
        "answer": """
        I know this:
        director: Christopher Nolan
        cast: Leonardo DiCaprio, Joseph Gordon-Levitt, Ellen Page, Tom Hardy
        budget: $160 million
        revenue: $829.9 million
        genre: Science Fiction, Thriller
        synopsis: A skilled thief is given a chance at redemption if he can successfully plant an idea in a target's subconscious.
        """
    },
    {
        "movie": "The Godfather",
        "answer": """
        I know this:
        director: Francis Ford Coppola
        cast: Marlon Brando, Al Pacino, James Caan, Diane Keaton
        budget: $6-7 million
        revenue: $250-287 million
        genre: Crime, Drama
        synopsis: The patriarch of an organized crime dynasty transfers control of his empire to his reluctant son.
        """
    },
    {
        "movie": "Avatar",
        "answer": """
        I know this:
        director: James Cameron
        cast: Sam Worthington, Zoe Saldana, Sigourney Weaver, Stephen Lang
        budget: $237 million
        revenue: $2.923 billion
        genre: Science Fiction, Adventure
        synopsis: A paraplegic Marine on an alien planet becomes torn between following his orders and protecting the world he feels is his home.
        """
    },
    {
        "movie": "Titanic",
        "answer": """
        I know this:
        director: James Cameron
        cast: Leonardo DiCaprio, Kate Winslet, Billy Zane, Kathy Bates
        budget: $200 million
        revenue: $2.243 billion
        genre: Romance, Drama
        synopsis: A young artist and a wealthy socialite fall in love aboard the ill-fated Titanic.
        """
    },
    {
        "movie": "The Dark Knight",
        "answer": """
        I know this:
        director: Christopher Nolan,
        cast: Christian Bale, Heath Ledger, Aaron Eckhart, Michael Caine
        budget: $185 million
        revenue: $1.006 billion
        genre: Action, Crime, Drama
        synopsis: Batman faces the Joker, a criminal mastermind who brings chaos and anarchy to Gotham City.
        """
    }
]

example_prompt = ChatPromptTemplate.from_messages(
    [
        ("human", "What do you know about {movie}?"),
        ("ai", "{answer}")
    ]
)

example_prompt = FewShotChatMessagePromptTemplate(
    example_prompt=example_prompt,
    examples=examples
)

final_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You possess extensive knowledge about movies. If there are any items you cannot answer, leave them blank without making anything up, and only respond to the questions asked."),
        example_prompt,
        ("human", "What do you know about {movie}?")
    ]
)

chain = final_prompt | llm

chain.invoke({"movie": "alien2"})
```
